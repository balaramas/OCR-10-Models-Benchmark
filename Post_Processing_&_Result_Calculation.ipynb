# Post Processing from MArker

```python
import pandas as pd
import re

markdown_pattern = r'[#*_`~>\-\+\[\]\(\)|!]'

def clean_markdown(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'\s+', ' ', text).strip()
    return re.sub(markdown_pattern, ' ', text).strip()

# Apply function to the specified column
df1["marker_txt"] = df1["marker_txt"].apply(clean_markdown)
```

```python
df2["marker_txt"] = df2["marker_txt"].apply(clean_markdown)
```

```python
df2["surya_txt"] = df2["surya_txt"].apply(clean_markdown)
```

```python
from os import uname
#

import jiwer
#scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)

def get_CER(text1, text2):
    if pd.isna(text2):
        text2 = ""
    cer = jiwer.cer(text1, text2)
    return cer


doctr_scores = []
pytes_scores = []
surya_scores = []
rapidocr_scores = []
google_vision_api_scores = []
paddleocr_scores = []
azure_intelligence_ai_scores = []
amazon_textract_scores = []
marker_scores = []
unstructured_io_scores = []

for index, row in df1.iterrows():
    #doctr_scores.append(get_f1_rougeL(row['funsd_txt'], row['doctr_txt']))
    pytes_scores.append(get_CER(row['funsd_txt'], row['pytes_txt']))
    surya_scores.append(get_CER(row['funsd_txt'], row['surya_txt']))
    rapidocr_scores.append(get_CER(row['funsd_txt'], row['rapidocr_txt']))
    google_vision_api_scores.append(get_CER(row['funsd_txt'], row['google_vision_api_txt']))
    paddleocr_scores.append(get_CER(row['funsd_txt'], row['paddleocr_txt']))
    azure_intelligence_ai_scores.append(get_CER(row['funsd_txt'], row['azure_intelligence_ai_txt']))
    amazon_textract_scores.append(get_CER(row['funsd_txt'], row['amazon_textract_api']))
    marker_scores.append(get_CER(row['funsd_txt'], row['marker_txt']))
    unstructured_io_scores.append(get_CER(row['funsd_txt'], row['unstructured_io_txt']))

#df['doctr_scores'] = doctr_scores

df = pd.DataFrame()
df['pytes_scores'] = pytes_scores
df['surya_scores'] = surya_scores
df['rapidocr_scores'] = rapidocr_scores
df['google_vision_api_scores'] = google_vision_api_scores
df['paddleocr_scores'] = paddleocr_scores
df['azure_intelligence_ai_scores'] = azure_intelligence_ai_scores
df['amazon_textract_scores'] = amazon_textract_scores
df['marker_scores'] = marker_scores
df['unstructured_io_scores'] = unstructured_io_scores

#df.head()
```
